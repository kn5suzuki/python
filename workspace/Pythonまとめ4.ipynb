# ライブラリ

## 標準ライブラリ
- pythonに標準で備わっているライブラリ
- いくつかのモジュールがまとまっている

## 外部ライブラリ
- サードパーティーが提供しており、pythonに標準には備わっていないためインストールする必要がある
- PyPIという場所でサードパーティライブラリが管理されており、十数万ものライブラリが登録されている

## pip
- パッケージ管理ツール
- コマンドプロンプトで実行できる
- PyPIからパッケージをインストールしたり、インストール済みのパッケージを確認したりできる

**コマンド例**

```python
!pip -V  #バージョンの確認
```

**インストール済みのライブラリ**
- listで見れる。**--outdated** で最新版でないライブラリのみ表示する
- freezeを使うことでバージョンとセットでライブラリの情報を得られる。**> txtファイル** とすることでテキストファイルに出力できる

```python
!pip list  #インストール済みのライブラリ
```

```python
!pip list --outdated
```

```python
!pip freeze
```

```python
!pip freeze > py_ver_2021_3.txt
```

**インストール**
- **ライブラリ==バージョン** でバージョン指定もできる
- **-r txtファイル** でテキストファイルのバージョンを復元できる
- **--upgrade** または **-U** でアップグレード。pip自身のアップグレードもできる

```python
!pip install --upgrade pip
```

```python
!pip install numpy  #インストール
```

```python
!pip install -U numpy
```

```python
!pip uninstall  #アンインストール
```

- コマンドの使い方の詳細は **-h** とすることで確認できる

```python
!pip install -h
```
