#### 3.3. plot
- 折れ線グラフを描く
- 引数は40個あり、**(x, y, fmt, scalex=True, scaley=True, data=None, キーワード引数)** のようになる。xは省略するとrange(len(y))となる
- fmtはcolor、marker、linestyleをショートカット文字列表記で指定したもの。キーワード引数と被った場合はキーワード引数が優先される
1. **label** : 文字列で凡例のラベルを指定
2. **color(c)** : 折れ線の色を指定（詳細は後述）
3. **dashes** : floatのシーケンスで実線部と空白部の長さを指定
4. **dash_capstyle** : デフォルトで**'butt'**。ダッシュ線の書式を指定。**'round','projecting'**
5. **dash_joinstyle** : デフォルトで**'miter'**。ダッシュ線の結合部分の書式を指定。**'round'、'bevel'**
6. **linestyle(ls)** : デフォルトでは**'-'**。折れ線の書式を指定。dashesを指定すると無効。**'-'、'- -'、'-.'、':'、'steps'**
7. **linewidth(lw)** : floatで折れ線の太さを指定
8. **alpha** : 0から1のfloatで透明度を指定
9. **zorder** : floatで優先度を指定。この値が大きいグラフが前面となる
10. **marker** : デフォルトで**'None'**。マーカーの形状を指定。**'.'、','、'o'、'v'、'^'、'<'、'>'、'1'、'2'、'3'、'4'、's'、'p'、'\*'、'h'、'H'、'+'、'x'、'D'、'd'、'|'、'_'**
11. **markersize(ms)** : floatでマーカーの大きさを指定
12. **markerfacecolor(mfc)** : マーカーの色
13. **markeredgewidth(mew)** : floatでマーカーの縁の幅を指定
14. **markeredgecolor(mec)** : マーカーの縁の色を指定

```python
y = [0, 1, 4, 9, 16, 25]  #最も基本的な描画
plt.plot(y)
plt.show()
```

```python
y = [0, 1, 4, 9, 16, 25]
plt.plot(y, label='line1', color='r', dashes=[2, 2, 10, 2], alpha=0.5, lw=5)  #dashes
plt.legend()
plt.show()
```

```python
y = [0, 1, 4, 9, 16, 25]
plt.plot(y, label='line1', color=[0, 1, 0], ls='-.', alpha=1, lw=1)  #ls
plt.legend()
plt.show()
```

```python
y = [0, 1, 4, 9,16, 25]
plt.plot(y, marker = 'o', ms = 10, mec = 'k', mew = 2)  #マーカー
plt.show()
```

```python
y = [0, 1, 4, 9,16, 25]
plt.plot(y, 'gd--')   #fmt
plt.show()
```

```python
y1 = [0, 1, 2, 3, 4, 5]    #グラフを重ねて書く
y2 = [0, 1, 4, 9, 16, 25]
plt.plot(y1, label='y=x')
plt.plot(y2, label='y=x^2')
plt.legend()
plt.show()
```

**色について**
- 色の指定の仕方には名前や16進数のカラーコードで指定する方法、1文字で指定する方法、RGBの割合で指定する方法がある
- 1文字で指定できるのは**b（青）、g（緑）、r（赤）、c（シアン）、m（マゼンタ）、y（黄色）、k（黒）、w（白）**
- matplotlibで指定されている色は **matplotlib.colors.cnames** で確認できる
- RGBで指定するときは0から1までのfloatで構成された長さ3のリストを用いる。[0, 0, 0]は黒となる

#### 3.3.2. title
- グラフにタイトルをつける
- **(label, fontdict=None, loc=None, pad=None)
- fontdictは辞書型でテキストの外観の設定をする
- locはデフォルトで'center'。'left', 'right'

#### 3.3.3. xlabel, ylabel
- グラフにラベルをつける
- 引数は **(label, fontdict=None, labelpad=None, loc=None)**

#### 3.3.4. xlim, ylim
- 軸の範囲を指定
- 引数は**(left, right)**またはそのタプル
- 戻り値はタプル

#### 3.3.5. grid
- グリッド線をつける
- 引数は **(b=None, which='major', axis='both')**
- bはグリッド線を引くかどうか。TrueかFalse
- whichはグリッド線の種類。'minor', 'both'
- axisはどの軸方向に引くか。'x', 'y'

```python
y = [0, 1, 4, 9,16, 25]
plt.plot(y)
plt.title('y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(0, 30)
plt.grid(True)
plt.show()
```

#### 3.4. scatter
- 散布図を描く
- 引数は **(x, y, キーワード引数)**
1. **s=None** : floatでマーカーのサイズを指定。デフォルトで20

6. **edgecolors=None** : 縁の色を指定

```python
height = [175, 169, 180, 172, 165, 176]
weight = [65, 60, 72, 67, 59, 70]
plt.scatter(height, weight)
plt.show()
```

#### 3.5. bar
- 棒グラフを描く
- 引数は **(x, y, キーワード引数)**
1. **width=0.8** : floatで棒の太さを指定
2. **bottom=None** : 棒の下側の余白、積み上げ棒グラフを作る時に利用
3. **align='center'** : 'edge'でバーの左端をxに合わせる
4. **color** : 棒の色
5. **edgecolor** : 縁の色
6. **linewidth** : 縁の幅
7. **tick_label** : リストでx軸ラベルを指定
8. **xerr** : 数値や配列でx軸方向のエラーバーを指定
9. **yerr** : y軸方向のエラーバー
10. **ecolor** : デフォルトで'k'。エラーバーの色
11. **capsize** : エラーバーの傘のサイズ
12. **log** : デフォルトでFalse。Trueで対数目盛

```python
x = range(5)
y = [2, 5, 3, 7, 1]
labels = ['A', 'B', 'C', 'D', 'E']
plt.bar(x, y, tick_label = labels)
plt.show()
```

```python
x = range(5)
y = [2, 5, 3, 7, 1]
labels = ['A', 'B', 'C', 'D', 'E']
plt.bar(x, y, tick_label=labels, width=0.3, color='c', edgecolor='y', linewidth=1)
plt.show()
```

```python
x = range(5)
y = [2, 5, 3, 7, 1]
yerr = [0.5, 0.6, 0.3, 0.8, 0.1]
labels = ['A', 'B', 'C', 'D', 'E']
plt.bar(x, y, tick_label=labels, xerr=0, yerr=yerr, capsize=5, ecolor='b')
plt.show()
```

```python
x = range(5)
y = [2, 50, 13, 70, 30]
labels = ['A', 'B', 'C', 'D', 'E']
plt.bar(x, y, tick_label = labels, log=True)
plt.show()
```

```python
x = range(5)
y1 = [2, 5, 3, 7, 1]
y2 = [1, 0.5, 2, 1.5, 1]
labels = ['A', 'B', 'C', 'D', 'E']
plt.bar(x, y1, tick_label=labels, bottom=y2)
plt.bar(x, y2)
plt.show()
```

#### 3.6. pie
- 円グラフを描く
- 引数は **(x, キーワード引数)**
1. **explode=None** : 配列で各要素を中心から離して目立たせる
2. **labels=None** : 配列で各リストのラベル
3. **colors=None** : 配列で各要素の色
5. **autopct=None** : 文字列で構成割合の表示方法を指定。一般的に**"%.〇f%%"** とする。%%は文字の%を意味している。
6. **pctdistance=0.6** : floatでパーセンテージの位置。0.0で中心。1.0で円周上
7. **shadow=False** : **True**で影を表示
4. **labeldistance=1.1** : ラベルの位置を指定
9. **startangle=0** : 出力の開始角度
8. **radius=1** : 円の半径
10. **counterclock=True** : **True**で時計回り。**False**で反時計回り
11. **wedgeprops=None** : 枠線の太さなど
12. **textprops=None** : 
13. **center=0** : 
14. **frame=False** : 
15. **rotatelabels=False** : 
16. **normalize=None** : 

```python
x = [13, 8, 5, 2, 1, 1]
labels = ['A', 'B', 'C', 'D', 'E', 'F']
plt.pie(x, labels=labels, autopct="%.1f%%", startangle = 90, counterclock=False)
plt.show()
```

```python
x = [13, 8, 5, 2, 1, 1]
labels = ['A', 'B', 'C', 'D', 'E', 'F']
c = [[0,0,1], [0,0.5,0.5], [0,1,0], [0,0.5,0], [0.5,0.5,0], [1,0,0]]
ex = [0.2, 0, 0, 0, 0, 0]
plt.pie(x, explode=ex, labels=labels, colors=c, autopct='%1.1f%%', startangle=90, counterclock=False)
plt.show()
```

#### 3.8.2. add_subplot
- 台紙だけでは出力できないので軸(Axes)をつける
- 引数は **(nrows, ncols, index)** で、それぞれ台紙の縦分割、横分割、インデックス。カンマを省略したpop引数でもよい

```python
fig = plt.figure(figsize=(2,3), dpi=72, facecolor='c', linewidth=1, edgecolor='k')
ax = fig.add_subplot(231)
ax = fig.add_subplot(236)
```

```python
fig = plt.figure(figsize=(2,3), dpi=72)
ax = fig.add_subplot(111)
y = [0, 1, 4, 9,16, 25]
ax.plot(y)
plt.show()
```

#### 3.8.3 subplots
- FigureとAxes(複数ならリスト)を返す
- 引数は **(nrows, ncols, キーワード引数)**

```python
fig, ax = plt.subplots(1,3)
print(ax)
x = [i for i in range(5)]
y = [i**2 for i in range(5)]
ax[0].plot(y)
ax[1].scatter(x,y)
plt.show()
```
