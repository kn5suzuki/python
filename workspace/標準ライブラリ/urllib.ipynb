# urllib

```python
import urllib
```

```python
response = urllib.request.urlopen('https://google.com/')
print(response.geturl())
```

```python
print(response.getcode())
```

```python
print(response.info())
```

```python
content = response.read()
print(content)
response.close()
```

```python
content.decode()  #引数でデコード方式の設定　デフォルトで'UTF-8'、その他'shift_jis'など
```
